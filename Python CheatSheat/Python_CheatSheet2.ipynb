```python
# Closures
# We have a closure in python when a nested function references a value of its enclosing function and then 
# the enclosing function returns the nested function

def get_multiplier(a):
    def out (b):
        return a * b
    return out
```

```python
multiply_by_3 = get_multiplier(3)
multiply_by_3(10)
```

```python
# If multiple nested functions within an enclosing function reference the same value, that value gets shared.
# To dynamically access a function's first free variable, use
'<function>._closure_[0].cell_contents'

# Example of nested functions sharing a variable (closure)

def outer ():
    x = 10                                # shared variable
    def inner1(): return x + 1
    def inner2(): return x + 2
    return inner1, inner2

f1, f2 = outer()
print(f1())    
```

```python
print(f2())
```

```python
# Now peak or modify shared variable dynamically

print(f1.__closure__[0].cell_contents)           # 10
f1.__closure__[0].cell_contents = 20                                     
```

```python
print(f1())
```

```python
print(f2())
```

```python
# Scope 
# If a variable is assigned to anywhere in the scope, it is regarded as a local variable, unless it is declared
# as 'global' or 'non local'

def get_counter():
    i = 0
    def out ():
        nonlocal i
        i += 1
        return i    
    return out
```

```python
# Debugger Example
# Decorator that prints function's name everytime it gets called

from functools import wraps

def debug(func):
    @wraps(func)
    def out (*args, **kwargs):
        print(func.__name__)
        return func (*args, **kwargs)
    return out

@debug
def add (x, y):
    return ( x + y)

# Call the decorated function to see the output
result = add(3,5)
print(f"Result: {result}")

# Wrap is a helper decorator that copies meta data of function add() to function out()
# Without it, 'add.__name__' would return 'out'

```

```python
# Inheritance

# Parent class, defines general attributes for a person
class Person:
    def __init__(self, name, age):
        self.name = name                 # instance's attribute: person's name
        self.age = age                   # instance's attribute, person's age

class Employee (Person):
    def __init__(self, name, age, staff_num):
        super().__init__(name, age)            # Call parent constructor to set name and age
        self.staff_num = staff_num             # instance attribute specific to Employee
    
```

```python
e1 = Employee('Bill', 28, 101)
e2 = Employee('Diamond', 30, 102)
print(e2.name)
```

```python
# Write text to file

def write_to_file(filename, text):
        with open (filename, 'w', encoding = 'utf-8') as file:
            file.write(text)
            
        # Read and write file    
        with open (filename, 'r', encoding = 'utf-8') as file:
            return file.readlines()                                # Return the lines from the file

# Write to file and then read and print each line
for line in write_to_file('myfile.txt', 'Liberian girl you came and you changed my world. Just like in the movies.'):
    print(line)
```

```python
# Append text to file

def append_to_file(filename, text):
    # first append the text to the file
    with open(filename, 'a', encoding = 'utf-8') as file:
        file.write(text)

    # Then read the file in a separate operation.
    with open (filename, 'r', encoding = 'utf-8') as file:
        return file.readlines()                                                                            

# Append to the file and print each line
for line in append_to_file('myfile.txt', '\n I love you Liberian girrrrrrrrrl, allll the time! \n'):
    print(line, end='')                  # end=''  removes extra newlines when printing
```

```python
# Match Object
import re
text = "Bill is friends with Bill and Billy."
pattern = r"Bill"

# Returns a match object, the first match
match = re.search(pattern, text)

# Access the matched string
if match:
    print(match.group())                     # 'Bill'

```

```python
import re

txt = 'John: 30, Alice: 24, Bob: 21'

# regex for capturing group name and age
pattern = r"(\w+): (\d+)"

# Find first match
match = re.search(pattern, txt)

if match:
    name = match.group(1)                       # Will give the first bracketed part  - 'John'
    all_parts = match.group()                   # All bracketed part - ('John', '30')
    start_idx = match.start()                   # Start index of match - 0
    end_idx = match.end()                       # Exclusive end index - 21

    print(" First captured parts:", name)
    print(" All captured parts:", all_parts)
    print(" Start index:", start_idx)
    print("End index:", end_idx)
    
    
```
